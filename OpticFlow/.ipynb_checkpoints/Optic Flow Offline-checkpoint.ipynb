# Eye description

```python


import numpy as np
import matplotlib.pylab as plt
import time 
%matplotlib inline

# date = np.str(time.strftime("%d/%m/%Y"))
# time = np.str(time.strftime("%I:%M:%S"))
# data_file = 'OFobjects_'+ date +'_'+ time+'.pickle'
```

```python
# Set eye parameters:

fov_x  = np.pi/2 #half FOV along azimuth in radians
fov_y = np.pi/4 # half FOV along elevation in radians
res_x = 41 #amount of pixels along azimuth
res_y = 21 # amount of pixels along elevation


```

```python
# Describe eye as a list of azimuth/elevation angles

phis   = np.linspace(-fov_x, fov_x, res_x)
thetas = np.linspace(-fov_y, fov_y, res_y)
layout = np.array([[phi, theta] 
                   for theta in thetas
                   for phi in phis])
print('Layout shape:', layout.shape)

# Display
plt.scatter(x=layout[:,0], 
            y=layout[:,1], 
            c=np.arange(layout.shape[0]))
```

# Unit vectors for each pixel
Equations for cartesian to spherical coordinates:
$$
u = \begin{bmatrix} 
        \cos(\theta) \cos(\phi) \\
        \cos(\theta) \sin(\phi) \\
        \sin(\theta)
\end{bmatrix}
$$

Where $\theta$ is the elevation angle and $\phi$ the azimuth angle


```python
# Compute unit vectors for each pixel
u = np.array([[ np.cos(theta) * np.cos(phi),
                np.cos(theta) * np.sin(phi),
                np.sin(theta)]
               for [phi, theta] in layout])
print('u shape:', u.shape)

# Function to display 3d vectors
def display_vectors(X, V):
    fig, axarr = plt.subplots(nrows=1, 
                              ncols=2, 
                              figsize=[10,5])
    # Top view
    for [ax, title, xlabel, ylabel, xx, yy, uu, vv] in zip(axarr, 
                                                   ['Top view', 'Side view'],
                                                   ['X', 'X'],
                                                   ['Y', 'Z'],
                                                   [X[0], X[0]],
                                                   [X[1], X[2]],
                                                   [V[0], V[0]],
                                                   [V[1], V[2]]):
        ax.set_title(title)
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        ax.set_xlim([-1, 1])
        ax.set_ylim([-1, 1])
        ax.quiver(xx, yy, uu, vv)

    fig.tight_layout()
    
# Display
display_vectors(u.T, u.T)
```

## Azimuth unit vector for each pixel

$$
a = \frac{z \times u}{||z \times u||}
$$

```python
# Compute a: lateral vector of the base (a, b, u)
z = np.array([0, 0, 1])
a = np.cross(z[np.newaxis], u)
a /= np.linalg.norm(a, axis=1)[:,np.newaxis]

# Display
display_vectors(u.T, a.T)
```

## Elevation unit vector for each pixel

$$
b = u \times a
$$

```python

# Compute b: vertical vector of the base (a, b, u)
b = np.cross(u, a)

# Display
display_vectors(u.T, b.T)
```
